CARP / CARL / PARL

```python
def pass_to_fileout(df):
    new_df = df[['user_id', 'item_id', 'ratings']].rename(columns={
                                                                "user_id":"UserId", 
                                                                "item_id":"ItemId",
                                                                "ratings":"Rating"
                                                                                    })
    new_df['Date'] = [i for i in range(len(new_df))]
    return new_df
```

```python
trainIterOut = pass_to_fileout(train)
trainIterOut.head()
```

```python
trainIterOut.to_csv("carp_carl_parl/TrainInteraction.out", sep="\t", header=None, index=False)
```

```python
ValIterOut = pass_to_fileout(val)
ValIterOut.head()
```

```python
ValIterOut.to_csv("carp_carl_parl/ValInteraction.out", sep="\t", header=None, index=False)
```

```python
testIterOut = pass_to_fileout(test)
testIterOut.head()
```

```python
testIterOut.to_csv("carp_carl_parl/TestInteraction.out", sep="\t", header=None, index=False)
```

```python
vocab = {}

i = 0
for text in data['reviews']:
    text = str(text).split()
    for word in text:
        if word not in vocab.keys():
            vocab[word] = i
            i += 1
```

```python
vocab = pd.DataFrame({
    "words":list(vocab.keys()),
    "id": list(vocab.values())
                      })
```

```python
vocab.to_csv("carp_carl_parl/WordDict.out", sep="\t", header=None, index=False)
```
